```python
from GoogleNews import GoogleNews
import pandas as pd

from IPython.display import display, HTML
```

```python
googlenews = GoogleNews(lang='pt', region='BR',period='1d')
googlenews.set_encode('utf-8')
```

```python
googlenews.clear()
#googlenews.search('Trabalho Escravo')
googlenews.search('Trabalho Escravo')
for i in range(2,3):
    googlenews.get_page(i)
    
```

```python
noticias = pd.DataFrame(googlenews.results())#sort=True\
```

```python
print("Manchete:",noticias['title'][0])
print("Mídia:",noticias['media'][0],noticias['date'][0])
print("link:", noticias['link'][0])
print(noticias['desc'][0])
```

```python
manchete = 0
print("- ",noticias['title'][manchete],"\n\t",noticias['media'][manchete],noticias['date'][manchete],"\n link:", noticias['link'][manchete])
```

```python
noticias
```

```python
lista_noticias=''
for i in range(0,noticias.shape[0]):
    lista_noticias = lista_noticias + '<li> <b>{}</b>  <br> &ensp; {}, {}. <a href="{}">leia mais</a></li>'.format(noticias['title'][i],\
                                noticias['media'][i],\
                                noticias['date'][i],\
                                noticias['link'][i])
```

```python
corpo = '''
    <h2>Aqui estão as últimas noticias sobre o tema</h2>
 
    <ul>
      {}
    </ul> 

'''.format(lista_noticias)
```
